### This is a sample Python notebook that constructs a shear force and bending moment diagram

```python
from sympy import *
from sympy.physics.continuum_mechanics.beam import Beam
import matplotlib.pyplot as plt

from matplotlib import style
style.use('ggplot')
init_printing()
```

```python
E, I = symbols('E, I')
b = Beam(10, E, I)  #beam of 10 units length with
                    #Young's modulus E and moment of inertia I
                    #about neutral axis
b.apply_load(-10, 5, -1) #load at midpoint, acting in negative Y direction
b.bc_deflection.append((0, 0))
#b.bc_slope.append((0, 0))        
b.bc_deflection.append((10, 0))
```

```python
R1, R2 = symbols('R1, R2')
b.apply_load(R1, 0, -1)    
b.apply_load(R2, 10, -1)
b.solve_for_reaction_loads(R1, R2)
b.reaction_loads
```
