```python
# Problem 1 at https://pages.mtu.edu/~dnaneet/2150/notes/problems/sfd_bmd_prelim.html
import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
E, I = sympy.symbols('E,I')
b = Beam(2, E, I)

b.apply_load(-1, 1, -1)
b.bc_deflection.append((0,0))
b.bc_deflection.append((2,0))

R1,R2 = sympy.symbols('R1, R2')
b.apply_load(R1, 0, -1)
b.apply_load(R2, 2, -1)

b.load
```

```python
b.solve_for_reaction_loads(R1, R2)
b.reaction_loads
```
